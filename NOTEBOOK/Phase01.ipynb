```python
import os
import shutil

base_dir = '/content/drive/MyDrive/datasets/skin_cancer9'

# Detect nested folder automatically
nested_folder = os.path.join(base_dir, os.listdir(base_dir)[0])
print("Nested folder detected:", nested_folder)

# Detect all subfolders in nested folder (actual splits)
splits = [f for f in os.listdir(nested_folder) if os.path.isdir(os.path.join(nested_folder, f))]
print("Detected split folders:", splits)

for split in splits:
    src_split = os.path.join(nested_folder, split)
    dst_split = os.path.join(base_dir, split.lower())  # normalize to lowercase
    os.makedirs(dst_split, exist_ok=True)
    for cls in os.listdir(src_split):
        src_cls = os.path.join(src_split, cls)
        dst_cls = os.path.join(dst_split, cls)
        shutil.move(src_cls, dst_cls)

# Remove the empty nested folder
shutil.rmtree(nested_folder)

# Verify flattening
print("Flattening done. Base folder contents now:")
print(os.listdir(base_dir))

# Verify each split has class folders
for split in os.listdir(base_dir):
    split_path = os.path.join(base_dir, split)
    if os.path.isdir(split_path):
        print(f"\nContents of {split}:")
        print(os.listdir(split_path))
```

```python
import os
import pandas as pd

splits = ['train','test']
rows = []

for split in splits:
    split_dir = os.path.join(base_dir, split)
    for cls in os.listdir(split_dir):
        cls_folder = os.path.join(split_dir, cls)
        n = 0
        for fn in os.listdir(cls_folder):
            if fn.lower().endswith(('.png','.jpg','.jpeg')):
                rows.append({'split': split, 'class': cls, 'path': os.path.join(cls_folder, fn)})
                n += 1
        print(f"{split} - {cls}: {n} images")

df = pd.DataFrame(rows)
manifest_path = os.path.join(base_dir,'manifest.csv')
df.to_csv(manifest_path, index=False)
print(f"\nManifest created with {len(df)} entries at {manifest_path}")
df.head()
```

```python
import os
import shutil
import random

base_dir = '/content/drive/MyDrive/datasets/skin_cancer9'
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'val')

# Create val folder
os.makedirs(val_dir, exist_ok=True)

val_ratio = 0.15  # 15% of training images for validation

# Process each class
for cls in os.listdir(train_dir):
    cls_train_folder = os.path.join(train_dir, cls)
    cls_val_folder = os.path.join(val_dir, cls)
    os.makedirs(cls_val_folder, exist_ok=True)

    files = [f for f in os.listdir(cls_train_folder) if f.lower().endswith(('.png','.jpg','.jpeg'))]
    n_val = int(len(files) * val_ratio)
    val_samples = random.sample(files, n_val)

    for f in val_samples:
        src = os.path.join(cls_train_folder, f)
        dst = os.path.join(cls_val_folder, f)
        shutil.move(src, dst)

print("Validation set created successfully.\n")

# Verify counts
for split in ['train','val','test']:
    print(f"\nContents of {split}:")
    split_dir = os.path.join(base_dir, split)
    for cls in os.listdir(split_dir):
        n = len([f for f in os.listdir(os.path.join(split_dir, cls)) if f.lower().endswith(('.png','.jpg','.jpeg'))])
        print(f"{cls}: {n} images")
```
