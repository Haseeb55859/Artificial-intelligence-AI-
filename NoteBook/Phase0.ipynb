# Phase 0.4 - Dataset Folder Structure

```python
# Phase 0.4 - Dataset Folder Structure
import os

# Base path inside your Drive
base_dir = '/content/drive/MyDrive/datasets/skin_cancer'

# Folder names
folders = [
    'train/NORMAL',
    'train/INFECTED',
    'val/NORMAL',
    'val/INFECTED',
    'test/NORMAL',
    'test/INFECTED'
]

# Create folders if not exist
for f in folders:
    path = os.path.join(base_dir, f)
    os.makedirs(path, exist_ok=True)

print("✅ Folder structure created at:", base_dir)
```

# Verification of Folders

```python
# Verification of Folders
for folder in ['train', 'val', 'test']:
    path = os.path.join(base_dir, folder)
    print(folder, "exists?", os.path.exists(path))
    if os.path.exists(path):
        print("  Subfolders:", os.listdir(path))
```
